```python
df = df.loc[df["year"] > 1989]
```

```python
offense_df = df.loc[(df["defense_sacks"] < 1) & (df["defense_tackles"] < 1) & (df["defense_tackle_assists"] < 1) & (df["defense_interceptions"] < 1) & (df["defense_interception_yards"] < 1) & (df["defense_interception_touchdowns"] < 1) & (df["defense_safeties"] < 1)]
```

```python
defense_columns = ['kick_return_attempts', 'kick_return_yards',
       'kick_return_touchdowns', 'punt_return_attempts', 'punt_return_yards',
       'punt_return_touchdowns', 'defense_sacks', 'defense_tackles',
       'defense_tackle_assists', 'defense_interceptions',
       'defense_interception_yards', 'defense_interception_touchdowns',
       'defense_safeties', 'punting_attempts',
       'punting_yards', 'punting_blocked']
```

```python
offense_df = offense_df.drop(defense_columns, axis=1)
```

```python
def score_calc(row):
    """Compute PPR score"""
    # starts from passing_yards, goes to field_goal_makes
    # we don't know how many yards they made the field goals from!!
    # weights_vec = np.array(0.04, 0, 4, -2, -0.1, 0, 0.1, 6, 0, 1, 0.1, 6, 0, 1, 3.4)
    ppr = 0
    ppr += 0.04 * row['passing_yards']
    ppr += 4 * row['passing_touchdowns']
    ppr += -2 * row['passing_interceptions']
    ppr += -0.1 * row['passing_sacks_yards_lost']
    ppr += 0.1 * row['rushing_yards']
    ppr += 6 * row['rushing_touchdowns']
    ppr += 1 * row['receiving_receptions']
    ppr += 0.1 * row['receiving_yards']
    ppr += 6 * row['receiving_touchdowns']
    ppr += 1 * row['point_after_makes']
    ppr += 3.4 * row['field_goal_makes']
    return ppr
```

```python
offense_df.agg([score_calc])
```
